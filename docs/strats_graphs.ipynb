```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("strat_benchmarks.csv")

order = ["ABPRUNING", "ABTABLE", "REGULARIDS", "tt_heavy", "killer_heavy", "cap_heavy"]
df["Strategy"] = pd.Categorical(df["Strategy"], categories=order, ordered=True)

pivot = df.pivot(index="Strategy", columns="Position", values="Time")

pivot.plot(kind="bar")

plt.title("Time Comparison Across Positions")
plt.ylabel("Time (s)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("time_comparison_all.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("strat_benchmarks.csv")

order = ["ABPRUNING", "ABTABLE", "REGULARIDS", "tt_heavy", "killer_heavy", "cap_heavy"]
df["Strategy"] = pd.Categorical(df["Strategy"], categories=order, ordered=True)

pivot = df.pivot(index="Strategy", columns="Position", values="Nodes")

pivot.plot(kind="bar")

plt.title("Nodes Comparison Across Positions")
plt.ylabel("Nodes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("nodes_comparison_all.png")
plt.show()
```

```python
pivot_tt = df.pivot(index="Strategy", columns="Position", values="TT")
pivot_tt.plot(kind="bar")

plt.title("Transposition Table Hits Comparison")
plt.ylabel("TT Hits")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("tt_hits_all.png")
plt.show()
```

```python
strategies = ["tt_heavy", "killer_heavy", "cap_heavy"]

filtered_df = df[df["Strategy"].isin(strategies)]

pivot_cut = filtered_df.pivot(index="Strategy", columns="Position", values="Cutoffs")
pivot_cut.plot(kind="bar")

plt.title("Cutoffs Comparison")
plt.ylabel("Cutoffs")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("cutoffs_all.png")
plt.show()
```
