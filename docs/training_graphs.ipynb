```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("weights_evolution_depth3.csv")
```

```python
plt.figure()

plt.plot(df["Generation"], df["Material"], label="Material")
plt.plot(df["Generation"], df["Safety"], label="Safety")
plt.plot(df["Generation"], df["Position"], label="Position")
plt.plot(df["Generation"], df["Control"], label="Control")

plt.title("Weights Evolution (Depth 3)")
plt.xlabel("Generation")
plt.ylabel("Weight Value")
plt.legend()
plt.grid()

plt.tight_layout()
plt.savefig("weights_evolution_depth3.png")
plt.show()
```
